```python
# Seasonal visualization of the data

# Convert kolom 'Date' menjadi datetime
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Ekstrak month dan season dari date
df['Month'] = df['Date'].dt.month

# Definisikan seasons: Summer (Dec-Feb), Autumn (Mar-May), Winter (Jun-Aug), Spring (Sep-Nov)
def get_season(month):
    if month in [12, 1, 2]:
        return 'Summer'
    elif month in [3, 4, 5]:
        return 'Autumn'
    elif month in [6, 7, 8]:
        return 'Winter'
    else:
        return 'Spring'

df['Season'] = df['Month'].apply(get_season)

# mengelompokkan data berdasarkan season dan hitung mean dari variabel yang relevan
seasonal_data = df.groupby('Season').agg({
    'MinTemp': 'mean',
    'MaxTemp': 'mean',
    'Rainfall': 'mean',
    'Sunshine': 'mean',
    'WindGustSpeed': 'mean'
}).reset_index()

# Sorting seasons
season_order = ['Summer', 'Autumn', 'Winter', 'Spring']
seasonal_data = seasonal_data.set_index('Season').loc[season_order].reset_index()

# Create subplots
plt.figure(figsize=(14, 10))

# Plot MinTemp by Season
plt.subplot(2, 2, 1)
plt.bar(seasonal_data['Season'], seasonal_data['MinTemp'], color='blue')
plt.title('Average MinTemp by Season')
plt.xlabel('Season')
plt.ylabel('MinTemp (°C)')

# Plot MaxTemp by Season
plt.subplot(2, 2, 2)
plt.bar(seasonal_data['Season'], seasonal_data['MaxTemp'], color='green')
plt.title('Average MaxTemp by Season')
plt.xlabel('Season')
plt.ylabel('MaxTemp (°C)')

# Plot Rainfall by Season
plt.subplot(2, 2, 3)
plt.bar(seasonal_data['Season'], seasonal_data['Rainfall'], color='cyan')
plt.title('Average Rainfall by Season')
plt.xlabel('Season')
plt.ylabel('Rainfall (mm)')

# Plot Sunshine by Season
plt.subplot(2, 2, 4)
plt.bar(seasonal_data['Season'], seasonal_data['Sunshine'], color='yellow')
plt.title('Average Sunshine by Season')
plt.xlabel('Season')
plt.ylabel('Sunshine (hours)')

plt.tight_layout()
plt.show()

```

Interpretasi:
1. Musim panas (Summer) memiliki suhu terhangat dengan curah hujan sedang dan sinar matahari yang melimpah.
2. Musim gugur (Autumn) memiliki curah hujan tertinggi, meskipun suhunya lebih moderat.
3. Musim dingin (Winter) adalah musim dengan suhu terendah dan hari-hari yang lebih pendek, tetapi curah hujan tetap tinggi.
4. Musim semi (Spring) menawarkan keseimbangan terbaik antara sinar matahari yang tinggi dan suhu yang moderat, dengan curah hujan yang lebih sedikit dibandingkan musim gugur dan musim dingin.

```python
# mengelompokkan data berdasarkan WindGustDir dan menghitung rata-rata Rainfall
rainfall_by_wind_direction = df.groupby('WindGustDir')['Rainfall'].mean().sort_values(ascending=False)

# Plotting the average rainfall by wind direction
plt.figure(figsize=(10, 6))
rainfall_by_wind_direction.plot(kind='bar', color='skyblue')
plt.title('Average Rainfall by Wind Direction')
plt.xlabel('Wind Gust Direction')
plt.ylabel('Average Rainfall (mm)')
plt.grid(True)
plt.show()
```

Dapat dilihat bahwa angin dari arah selatan (SE, SSE, SSW, dan SW) cenderung membawa curah hujan yang lebih tinggi dibandingkan angin dari arah utara (N, NNE, NNW). Juga, arah angin dari timur (ESE dan E) juga memiliki kontribusi besar terhadap curah hujan, sedangkan angin dari barat dan barat laut (W, WNW, NW) membawa curah hujan yang lebih rendah.


```python
# top 25% wind gust speed threshold
wind_threshold = df['WindGustSpeed'].quantile(0.75)

# Klasifikasi cuaca menjadi 'Windy' dan 'Calm' berdasarkan threshold
df['WindCategory'] = df['WindGustSpeed'].apply(lambda x: 'Windy' if x >= wind_threshold else 'Calm')

# Grouping berdasarkan WindCategory dan mencari mean dari variabel cuaca lainnya
windy_vs_calm = df.groupby('WindCategory').agg({
    'MinTemp': 'mean',
    'MaxTemp': 'mean',
    'Rainfall': 'mean',
    'Sunshine': 'mean',
    'WindGustSpeed': 'mean'
}).reset_index()

windy_vs_calm

plt.figure(figsize=(14, 8))

# Plot MinTemp
plt.subplot(2, 2, 1)
plt.bar(windy_vs_calm['WindCategory'], windy_vs_calm['MinTemp'], color='blue')
plt.title('Average MinTemp: Windy vs Calm Days')
plt.xlabel('Day Type')
plt.ylabel('MinTemp (°C)')

# Plot MaxTemp
plt.subplot(2, 2, 2)
plt.bar(windy_vs_calm['WindCategory'], windy_vs_calm['MaxTemp'], color='green')
plt.title('Average MaxTemp: Windy vs Calm Days')
plt.xlabel('Day Type')
plt.ylabel('MaxTemp (°C)')

# Plot Rainfall
plt.subplot(2, 2, 3)
plt.bar(windy_vs_calm['WindCategory'], windy_vs_calm['Rainfall'], color='cyan')
plt.title('Average Rainfall: Windy vs Calm Days')
plt.xlabel('Day Type')
plt.ylabel('Rainfall (mm)')

# Plot Sunshine
plt.subplot(2, 2, 4)
plt.bar(windy_vs_calm['WindCategory'], windy_vs_calm['Sunshine'], color='yellow')
plt.title('Average Sunshine: Windy vs Calm Days')
plt.xlabel('Day Type')
plt.ylabel('Sunshine (hours)')

plt.tight_layout()
plt.show()
```

1. Suhu (baik minimum maupun maksimum) tidak banyak berbeda antara hari berangin dan hari tenang.
2. Curah hujan cenderung lebih tinggi pada hari berangin, menunjukkan bahwa kondisi berangin lebih sering dikaitkan dengan cuaca basah.
3. Sinar matahari sedikit lebih banyak pada hari tenang, menunjukkan bahwa hari tenang cenderung lebih cerah dibandingkan dengan hari berangin.

```python
# Klasifikasi hari menjadi 'Dry' dan 'Wet'
df['RainCategory'] = df['Rainfall'].apply(lambda x: 'Wet' if x > 0 else 'Dry')

# Grouping berdasarkan RainCategory dan menghitung mean dari tiap variabel cuaca lainnya
dry_vs_wet = df.groupby('RainCategory').agg({
    'MinTemp': 'mean',
    'MaxTemp': 'mean',
    'WindGustSpeed': 'mean',
    'Sunshine': 'mean',
    'Rainfall': 'mean'
}).reset_index()

# Create subplots
plt.figure(figsize=(14, 8))

# Plot MinTemp
plt.subplot(2, 2, 1)
plt.bar(dry_vs_wet['RainCategory'], dry_vs_wet['MinTemp'], color='blue')
plt.title('Average MinTemp: Dry vs Wet Days')
plt.xlabel('Day Type')
plt.ylabel('MinTemp (°C)')

# Plot MaxTemp
plt.subplot(2, 2, 2)
plt.bar(dry_vs_wet['RainCategory'], dry_vs_wet['MaxTemp'], color='green')
plt.title('Average MaxTemp: Dry vs Wet Days')
plt.xlabel('Day Type')
plt.ylabel('MaxTemp (°C)')

# Plot WindGustSpeed
plt.subplot(2, 2, 3)
plt.bar(dry_vs_wet['RainCategory'], dry_vs_wet['WindGustSpeed'], color='purple')
plt.title('Average WindGustSpeed: Dry vs Wet Days')
plt.xlabel('Day Type')
plt.ylabel('WindGustSpeed (km/h)')

# Plot Sunshine
plt.subplot(2, 2, 4)
plt.bar(dry_vs_wet['RainCategory'], dry_vs_wet['Sunshine'], color='yellow')
plt.title('Average Sunshine: Dry vs Wet Days')
plt.xlabel('Day Type')
plt.ylabel('Sunshine (hours)')

plt.tight_layout()
plt.show()
```

1. Suhu: Hari basah cenderung memiliki suhu maksimum yang sedikit lebih rendah daripada hari kering, tetapi suhu minimum tetap sama.
2. Kecepatan angin: Tidak ada perbedaan berarti antara hari kering dan basah dalam hal kecepatan angin terkuat.
3. Sinar matahari: Hari kering lebih cerah dengan lebih banyak jam sinar matahari, sementara hari basah cenderung lebih mendung dengan sinar matahari yang lebih sedikit.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
# Assuming your dataset is loaded into a variable named 'df'

# Filter only numerical columns
numerical_columns = ['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine', 'WindGustSpeed']

# Create pairplot with linear regression line
sns.set(style="whitegrid")
pairplot = sns.pairplot(df[numerical_columns], kind="reg", diag_kind="kde", markers="+", plot_kws={'line_kws':{'color':'red'}, 'scatter_kws': {'alpha': 0.5}})

# Adding correlation coefficients
def annotate_correlation(x, y, **kwargs):
    corr = np.corrcoef(x, y)[0, 1]
    ax = plt.gca()
    ax.annotate(f'Corr: {corr:.2f}', xy=(.5, .9), xycoords=ax.transAxes, ha='center', fontsize=12)

pairplot.map(annotate_correlation)

# Show plot
plt.show()
```

Dari visualisasi tersebut didapatkan hasil dari relationship masing-masing variabel secara linear.

Informasi yang dapat diambil dari hal ini adalah:


*   Terdapat hubungan linear positif yang kuat antara variabel MinTemp dan MaxTemp. Hal ini masuk akal karena pada dasarnya, kedua variabel tersebut adalah variabel yang saling berkaitan.
*   Terdapat hubungan linear positif yang cukup kuat dengan nilai korelasi 0.57 antara MinTemp dengan Evaporation. Hal ini dapat mengindikasikan bahwa suhu minimum dalam celcius memiliki hubungan linear yang cukup kuat dengan penguapan (evaporasi).
* Terdapat hubungan linear positif yang cukup kuat dengan nilai korelasi 0.51 antara MaxTemp dengan Evaporation. Hal ini dapat mengindikasikan bahwa suhu maksimum dalam celcius memiliki hubungan linear yang cukup kuat dengan penguapan (evaporasi).

Untuk keterhubungan linear antara masing-masing variabel dengan variabel lainnya satu sama lain, tidak mengindikasikan adanya keterhubungan linear yang kuat lagi. Sehingga dari sini dapat diambil informasi bahwa untuk proses penguapan atau evaporasi, hal yang paling berkorelasinya secara linear adalah suhu atau temperature.

```python
# Menghitung korelasi antar variabel
df_numerical = df[numerical_columns]
correlation_matrix = df_numerical.corr()

# Membuat heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1, linewidths=0.5)

# Menambahkan judul
plt.title("Heatmap Korelasi Antar Variabel", fontsize=16)

# Menampilkan plot
plt.show()
```

Hasil tersebut adalah hasil heatmap korelasi linear antara masing-masing variabel. Didapatkan informasi yang hampir sama seperti pada bagian sebelumnya.
